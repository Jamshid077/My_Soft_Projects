```python
print(df.groupby('genre').size().sort_values(ascending=False))
```

```python
print(df.groupby('platform').size().sort_values(ascending=False))
```

```python
df['publisher']=df['publisher'].apply(lambda x:x.strip())
df['publisher']=df['publisher'].apply(lambda x:x.title())
print(df.groupby('publisher').size().sort_values(ascending=False))
```

```python
df['developer']=df['developer'].apply(lambda x:x.strip())

df['developer']=df['developer'].apply(lambda x:x.title())

print(df.groupby('developer').size().sort_values(ascending=False))
```
